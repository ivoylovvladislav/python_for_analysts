```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', encoding='ISO-8859-1')
df.head()
```

### 1.1 
Создать новый признак Cpu_Company, который будет
содержать только название фирмы, которая произвела
CPU

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.head()
```

### 1.2 
Создать новый признак Memory_Amount, который будет
содержать только количество Gb памяти без указания типа
носителя


```python
def convert_TB_to_GB(string: str) -> int:
    new_str = string.split()[0]
    if new_str.endswith('TB'):
        result = float(new_str.replace('TB', '')) * 1024
    else: 
        result = new_str.replace('GB', '')
    return int(result)
```

```python
df['Memory_Amount'] = df['Memory'].apply(convert_TB_to_GB)
df.head(30)
```

### 1.3 
Создать новый признак Memory_Type, который будет
содержать только тип носителя (HDD/SDD/др.)

```python
df['Memory_Type'] = df['Memory'].apply(lambda x: x.split()[1])
df.head()
```

### 1.4 
Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
```

```python
df.head()
```

### 2.1 
Создайте признак SSD, который изначально равен 0

```python
df['SSD'] = 0
df.head()
```

### 2.2 
Поставьте в признаке SSD 1, если ноутбук
действительно с типом носителя SSD


```python
df.loc[df['Memory_Type'] == 'SSD', 'SSD'] = 1
df.head()
```

### 2.3 
Уберите в признаке Weight значения 'kg' и поменяйте его
тип данных на вещественный


```python
df['Weight'] = df['Weight'].str.replace('kg', '')
df.head()
```

```python
df['Weight'] = df['Weight'].astype('float')
df.info()
```
